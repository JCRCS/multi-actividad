# 1. Explorando bases de datos y sus tablas
Comandos de exploración de bases de datos<br />
show databases: Lista las bases de datos disponibles<br />
use database: Cambiarse de base de datos<br />
show tables in database: Lista las tablas disponibles en la base de datos especificada<br />
show tables like 'wildcard_pattern':Lista las tablas disponibles en la bae de datos actual cuyo nombre cumple con el patrón especificado<br />
describe table: Muestra un reporte con la lista de campos de la tabla y sus tipos de datos<br />
show table stats table: Muestra las estadísticas de la tabla<br />
show partitions table: Lista las particiones de la tabla<br />
show files in table: Lista los archivos físicos que componene una tabla<br />
